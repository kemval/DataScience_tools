# Exploratory Data Analysis (EDA) for Housing Data
This notebook performs Exploratory Data Analysis (EDA) on a housing dataset to understand its characteristics and prepare it for further analysis or modeling. We'll explore missing values, identify categorical features, analyze numerical features, and examine relationships between variables.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('housing_price_dataset.csv')
print(data.head())
print(data.dtypes)
```

## 1. Missing Value Analysis

Description: This section checks for missing values in the dataset. We'll use the .isnull().sum() method of pandas to identify any missing entries. Missing values can affect analysis, so it's crucial to address them before proceeding.

```python
print("Missing values:")
print(data.isnull().sum())
```

## 2. Identifying Categorical Features

Description: This section explores the categorical variables present in the data. Categorical variables represent qualities or classifications that don't have a natural order (e.g., Neighborhood: Rural, Suburb, Urban). We'll use the .select_dtypes(include='object') method of pandas to identify these features.

```python
categorical_features = data.select_dtypes(include=['object'])
print("Categorical features:")
print(categorical_features.columns.tolist())
```

## 3. Exploring Numerical Features (Optional)

Description: While not directly printing the result here, this section acknowledges the creation of a DataFrame containing only the numerical features using .select_dtypes(exclude='object'). This DataFrame will be used for further analysis in the following sections.

```python
numerical_features = data.select_dtypes(exclude=['object'])
print(numerical_features.columns.tolist())
```

## 4. Outlier Detection with Boxplots

Description: This section utilizes boxplots to visualize the distribution of data points in each numerical feature. Boxplots are helpful for identifying potential outliers, which are data points that fall significantly outside the main cluster of data. Outliers can skew analysis, so it's important to be aware of their presence.

```python
for col in numerical_features.columns:
    sns.boxplot(x = col, data=data)
    plt.title(f"Boxplot for {col}")
    plt.show()
```

## 5. Distribution of Categorical Feature: Neighborhood

Description: This section analyzes the distribution of houses across different neighborhoods (Rural, Suburb, Urban, etc.) using countplots. Countplots visualize the frequency of each category in a categorical variable. Understanding the distribution helps identify any biases or imbalances in the data regarding neighborhood types.

```python
for col in categorical_features.columns:
    sns.countplot(x = col, data=data)
    plt.title(f"Distribution of {col}")
    plt.show()
```

## 6. Relationship between Numerical Features and Price

Description: This section explores the relationships between numerical features (e.g., Square Footage, Bedrooms) and the target variable (likely 'Price') using scatter plots. Scatter plots visually depict how two variables change together. We'll discuss any observed trends or correlations between features and price, which can be helpful for understanding how different factors might influence housing prices.

```python
for col in numerical_features.columns:
    sns.scatterplot(x = col, y='Price', data=data)
    plt.title(f"Relationship between {col} and Price")
    plt.show()
```
