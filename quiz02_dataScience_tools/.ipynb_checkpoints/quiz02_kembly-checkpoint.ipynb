# Salary Analysis in the Data Science Field

## Introduction

This Jupyter Notebook is dedicated to performing an exploratory data analysis on a dataset containing salary information within the data science domain. The analysis focuses on understanding salary distributions, the impact of job categories, remote work preferences, and the influence of employment types on salaries.

The goal is to derive meaningful insights that can guide the organization's salary policies and hiring strategies, ensuring competitive and fair compensation practices.


## 1. Data Loading and Overview

The first step in my exploratory data analysis is to load the dataset into a Pandas DataFrame. This process involves reading the data from a CSV file into a structured format that allows for easy manipulation and analysis with Python.

### Dataset Description

The dataset, named `ds_salaries.csv`, contains salary information within the data science domain. It includes various attributes such as work year, experience level, employment type, job title, salary in both local currency and USD, employee residence, remote work ratio, company location, and company size. This rich set of features will enable us to explore various aspects of salary distributions and factors influencing compensation in the data science field.

### Objectives of the Overview

- **Understand the Structure:** By displaying the first few rows of the DataFrame, we aim to get a preliminary understanding of the dataset's structure, including the types of data (numerical, categorical) it contains and how they are formatted.

- **Identify Key Columns:** Initial observation helps in identifying the key columns that will be the focus of our analysis, such as `salary_in_usd`, `job_title`, `experience_level`, etc.

- **Preliminary Observations:** Make some initial observations that could guide the direction of our detailed analysis. This may include noting any apparent inconsistencies, missing values, or interesting trends that warrant a closer look.

### Loading the Data

The code snippet below demonstrates how to load the dataset using Pandas, a powerful and flexible data analysis library in Python. We then display the first few rows to start our exploration.



```python
import pandas as pd

df = pd.read_csv('ds_salaries.csv')
df.head()
```

## 2. Data Cleaning

After loading our dataset, the next critical step is to clean the data. Data cleaning is essential to ensure the accuracy of our analysis. This process involves handling missing values, removing duplicate entries, and ensuring that each column has the appropriate data type for its contents. Let's delve into each of these steps:

### Handling Missing Values

Missing data can significantly impact the conclusions drawn from the analysis. There are several strategies for handling missing values, including:

- **Removing Rows or Columns:** If a column or row contains a high percentage of missing values, it might be more beneficial to remove them entirely.
- **Imputation:** For columns with a significant amount of useful data and some missing values, we can fill in (or impute) the missing values with a summary statistic (mean, median, mode) or use more complex methods based on other columns.

The choice of strategy depends on the context and the proportion of missing data. We aim to balance data integrity with the need for a complete dataset.

### Removing Duplicates

Duplicate entries can skew our analysis, leading to incorrect conclusions. Identifying and removing duplicates ensures each data point is unique, contributing to the reliability of our findings.

### Ensuring Consistent Data Types

Data types are crucial for correct data processing. For example, numerical columns should have numeric types (`int`, `float`), and categorical data should be of type `object` or `category`. Correct data types allow for appropriate aggregations and ensure computational efficiency.

- **Numeric Columns:** We check that columns representing numerical values (e.g., salary) are in numeric data types.
- **Categorical Columns:** Columns representing categories (e.g., job titles, company sizes) should be in the `category` data type if they have a limited set of values, enhancing memory usage and performance.

### Applying Data Cleaning Steps

Below are the code snippets corresponding to each data cleaning step described above. Note that the specific actions might vary depending on the initial assessment of the dataset's condition.



```python
# Check for missing values
missing_values = df.isnull().sum()
print("Missing values in each column:\n", missing_values)

# Handling missing values
df_cleaned = df.dropna()
```

## 3. Descriptive Statistics

A crucial step in exploratory data analysis is to generate descriptive statistics and explore data distributions. This allows us to grasp the central tendencies, variability, and shape of our dataset's distributions at a glance. In this section, we focus on:

### Generating Summary Statistics for Numerical Columns

Summary statistics provide a quick overview of the measures of central tendency (such as mean and median), dispersion (such as range, interquartile range, standard deviation), and shape (skewness, kurtosis) of the numerical columns in our dataset. These metrics are fundamental for understanding the data's characteristics before diving deeper into specific analyses.

- **Mean** and **Median** offer insights into the central tendency of the data.
- **Standard Deviation** and **Interquartile Range (IQR)** inform us about the spread of the data points around the central value.
- **Min** and **Max** values highlight the range within which the data points fall.
- **Count** gives us the total number of non-null values in each column, which is useful for identifying missing data.

### Exploring the Distribution of Salaries

Salaries, both in local currency and in USD, are of particular interest in our dataset. Exploring these distributions can provide insights into compensation trends within the data science field, including:

- **Salary Ranges**: Understanding the spread of salaries can help in identifying the range within which most salaries fall, as well as spotting outliers.
- **Local vs. USD Currency**: Analyzing salaries in local currencies alongside their USD equivalents allows for a comparison of compensation trends across different geographical locations.
- **Distribution Shape**: Visualizing the salary distribution can reveal if the data is skewed, normal, or has any peculiar characteristics that warrant further investigation.

In the following sections, we'll generate summary statistics for our numerical columns and visually explore the salary distributions to glean insights from our dataset.




```python
# Generate summary statistics for numerical columns
summary_statistics = df_cleaned.describe()
print(summary_statistics)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Setting seaborn plots
sns.set(style="whitegrid")

# Histogram of salaries in USD 
plt.figure(figsize=(10, 6))
sns.histplot(df_cleaned['salary_in_usd'], kde=True, color='#E75480')
plt.title('Distribution of Salaries in USD')
plt.xlabel('Salary (USD)')
plt.ylabel('Frequency')
plt.show()

# Boxplot of salaries in USD 
plt.figure(figsize=(10, 6))
sns.boxplot(x=df_cleaned['salary_in_usd'], color='#E75480')
plt.title('Boxplot of Salaries in USD')
plt.xlabel('Salary (USD)')
plt.show()
```

## 4. Experience-Level Analysis

A key factor influencing salary in any field, including data science, is the level of experience. Understanding how experience impacts salary can provide valuable insights into career progression, compensation fairness, and hiring practices. In this section, we aim to:

### Analyze the Average Salary by Experience Level

By calculating the average salary associated with each experience level, we can identify trends and disparities in how experience translates into compensation. This analysis helps in answering questions such as:

- Do more experienced professionals consistently earn higher salaries?
- What is the salary progression like from entry-level to senior positions?

### Visualize the Distribution of Experience Levels

Exploring the composition of experience levels within our organization gives us an overview of the workforce's maturity and diversity. It can also highlight potential gaps or imbalances in experience that may need addressing. Questions of interest include:

- What proportion of our workforce is at an entry-level versus a senior level?
- Are there sufficient mid-level professionals to bridge the gap between entry-level and senior roles?

### Methodology

We will first group our dataset by experience level and calculate the average salary for each group. Then, we will create a visualization to display the distribution of experience levels within the organization, providing a clear picture of our workforce composition.

Following this analysis, we can draw insights into the relationship between experience and salary within our dataset and make informed recommendations for addressing any identified issues.


```python
average_salary_by_experience = df_cleaned.groupby('experience_level')['salary_in_usd'].mean().reset_index()
average_salary_by_experience.columns = ['Experience Level', 'Average Salary (USD)']
print(average_salary_by_experience)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

#color palette
pink_palette = ["#FFC0CB", "#FF69B4", "#E75480"] 

# Setting the aesthetics for seaborn plots
sns.set(style="whitegrid")

# Visualize the distribution of experience levels
plt.figure(figsize=(10, 6))
sns.countplot(x='experience_level', data=df_cleaned, palette=pink_palette, order=df_cleaned['experience_level'].value_counts().index)
plt.title('Distribution of Experience Levels within the Organization')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

## 5. Employment-Type Trends Analysis

The nature of employment, whether full-time, part-time, or contract, plays a significant role in determining salary structures and work arrangements within organizations. By investigating how these employment types influence salary trends and remote work preferences, we can gain insights into the evolving landscape of work and compensation in the data science field. This section is dedicated to:

### Analyzing Salary Trends by Employment Type

Salaries can vary significantly based on employment type due to differences in work hours, job responsibilities, and contract terms. Understanding these variations helps in assessing the equity and competitiveness of salaries offered to employees in different employment arrangements. Key questions include:

- How do average salaries compare between full-time, part-time, and contract employees?
- Are certain employment types more lucrative on average than others?

### Comparing Remote Work Ratios by Employment Type

The flexibility to work remotely has become an important aspect of job satisfaction and work-life balance. Examining how remote work preferences or possibilities vary among different employment types can reveal trends in work arrangements and employee expectations. We aim to explore:

- Which employment types are more likely to offer remote work options?
- How does the possibility of remote work correlate with employment type?

### Methodology

This analysis will enable us to draw conclusions about the relationship between employment type, salary, and remote work, offering valuable insights for organizational policy-making and strategy development.

```python
# Calculate the remote work ratio by employment type
remote_work_ratio_by_employment_type = df_cleaned.groupby('employment_type')['remote_ratio'].mean().reset_index()
remote_work_ratio_by_employment_type.columns = ['Employment Type', 'Average Remote Ratio']
print(remote_work_ratio_by_employment_type)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


#color palette
pink_palette = ["#FFC0CB", "#FF69B4", "#E75480"] 

# Setting the aesthetics for seaborn plots
sns.set(style="whitegrid")

# Bar plot for average salary by employment type
plt.figure(figsize=(10, 6))
sns.barplot(x='Employment Type', y='Average Salary (USD)', data=average_salary_by_employment_type, palette=pink_palette)
plt.title('Average Salary by Employment Type')
plt.show()

# Bar plot for average remote work ratio by employment type
plt.figure(figsize=(10, 6))
sns.barplot(x='Employment Type', y='Average Remote Ratio', data=remote_work_ratio_by_employment_type, palette=pink_palette)
plt.title('Average Remote Work Ratio by Employment Type')
plt.show()
```

## 10. Job Title Formatting and Categorization

The `job_title` column in our dataset contains a wide range of titles, reflecting the diversity of roles within the data science domain. However, for analytical purposes, it's beneficial to categorize these varied titles into broader groups. This categorization enables us to perform more streamlined and focused analyses on role-specific trends and insights. This section outlines the steps to:

### Identify Unique Job Titles

Initially, we'll explore the unique job titles present in our dataset to understand the variety and specificity of roles captured in the data.

### Create a Categorization Function

To simplify our analysis, we'll create a function that maps the diverse job titles into predefined categories. These categories are designed to encompass the primary roles within the data science field, including:

- **Data Scientist:** For all variations of data scientist titles.
- **Machine Learning Engineer:** For titles specifically mentioning machine learning.
- **Data Analyst:** For titles identifying the role as focusing on data analysis.
- **Data Engineer:** For titles related to data engineering.
- **Business Intelligence and Analytics:** For titles related to BI and analytics.
- **Other:** For titles that clearly fit into other recognized categories but are still relevant to the field.
- **Uncategorized:** For titles that don't clearly fit into any of the above categories.

### Apply the Categorization Function

With the function defined, we'll apply it across the `job_title` column to replace the original titles with the standardized categories, facilitating a more consolidated and focused analysis.

### Methodology

The process involves using the `apply()` method in Pandas to transform the `job_title` column based on the logic defined in our categorization function. This approach ensures consistency and clarity in how we refer to and analyze different roles within the dataset.


```python
# Identify unique job titles
unique_job_titles = df_cleaned['job_title'].unique()
print("Unique Job Titles:", unique_job_titles)
```

```python
def categorize_job_title(job_title):
    job_title_lower = job_title.lower()
    if 'data scientist' in job_title_lower or 'data science' in job_title_lower:
        return "Data Scientist"
    elif 'machine learning' in job_title_lower or 'ml' in job_title_lower or 'deep learning' in job_title_lower or 'ai' in job_title_lower:
        return "Machine Learning Engineer"
    elif 'data analyst' in job_title_lower or 'analytics' in job_title_lower or 'data analytics' in job_title_lower or 'insight' in job_title_lower:
        return "Data Analyst"
    elif 'data engineer' in job_title_lower or 'etl' in job_title_lower or 'data devops' in job_title_lower or 'data infrastructure' in job_title_lower or 'data operations' in job_title_lower:
        return "Data Engineer"
    elif 'business intelligence' in job_title_lower or 'bi ' in job_title_lower:
        return "Business Intelligence and Analytics"
    elif 'data architect' in job_title_lower or 'data strategy' in job_title_lower or 'data management' in job_title_lower or 'data quality' in job_title_lower:
        return "Data Architect/Manager"
    elif 'research' in job_title_lower or 'nlp' in job_title_lower or 'natural language processing' in job_title_lower or 'computer vision' in job_title_lower:
        return "Research Scientist/Engineer"
    elif 'head of data' in job_title_lower or 'director of data science' in job_title_lower or 'data science manager' in job_title_lower:
        return "Data Leadership"
    else:
        return "Other"

# Reapply the modified categorization function to the job_title column
df_cleaned['job_category'] = df_cleaned['job_title'].apply(categorize_job_title)

# Display the first few rows to verify the updated categorization
df_cleaned[['job_title', 'job_category']].head()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Setting the aesthetics for seaborn plots
sns.set(style="whitegrid")

# pink palette
pink_palette = [
    "#FADADD",  
    "#F4C2C2",  
    "#FFB6C1",  
    "#FF69B4",  
    "#C71585",  
    "#DB7093",  
    "#FF1493",  
    "#FFC0CB",
    "#FF69B4"
]

# Ensure the palette has enough colors for each category
unique_categories = df_cleaned['job_category'].nunique()
if len(pink_palette) < unique_categories:
    print(f"Warning: You have {unique_categories} unique categories but only {len(pink_palette)} colors in your palette. Consider expanding the palette.")

# Visualize the distribution of job categories
plt.figure(figsize=(12, 6))
sns.countplot(y='job_category', data=df_cleaned, 
              order=df_cleaned['job_category'].value_counts().index,  
              palette=pink_palette)
plt.title('Distribution of Job Categories')
plt.xlabel('Count')
plt.ylabel('Job Category')
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Expanded custom pink palette
pink_palette_expanded = [
    "#FFC0CB",  
    "#FFB6C1",  
    "#FF69B4",  
    "#FF1493",  
    "#C71585",  
    "#DB7093",  
    "#E75480",  
    "#D1479F",
    "#DB7093"
]

# Ensure the palette has enough colors for each category
unique_categories = df_cleaned['job_category'].nunique()
if len(pink_palette_expanded) < unique_categories:
    print(f"Warning: You have {unique_categories} unique categories but only {len(pink_palette_expanded)} colors in your palette. Consider expanding the palette or repeating some colors.")

# Boxplot of salaries by job category using the expanded pink palette
plt.figure(figsize=(12, 8))
sns.boxplot(x='salary_in_usd', y='job_category', data=df_cleaned,
            order=df_cleaned['job_category'].value_counts().index,
            palette=pink_palette_expanded)
plt.title('Salary Distribution by Job Category')
plt.xlabel('Salary (USD)')
plt.ylabel('Job Category')
plt.show()

```
